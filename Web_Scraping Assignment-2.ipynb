# 7 iphone11_Review WebScraping

```python
#importing all the necessary libraries
import selenium
from selenium import webdriver
import time
```

```python
#here we are connecting to the webdriver
```

```python
driver = webdriver.Chrome(r"C:\Users\91981\Desktop\chromedriver_win32/chromedriver.exe")
```

```python
#here specifying the url of the webpage to be scrapped
```

```python
driver.get("https://www.flipkart.com/apple-iphone-11-black-64-gb-includes-earpods-power-adapter/product-reviews/itm0f37c2240b217?pid=MOBFKCTSVZAXUHGR&lid=LSTMOBFKCTSVZAXUHGR3QP11A&marketplace=FLIPKART")
```

```python
#here we are creating the empty list to create a DataFrame
```

```python
Rating_phone=[]
Review_summary=[]
Full_review=[]
```

```python
#Now we aregoing to scrap the data 
```

```python
for i in range (0 ,10):
    for i in driver.find_elements_by_xpath("//div[@class='_3LWZlK _1BLPMq']"):
        Rating_phone.append(i.text)
    for m in driver.find_elements_by_xpath("//p[@class='_2-N8zT']"):
        Review_summary.append(m.text)
    for n in driver.find_elements_by_xpath("//div[@class='t-ZTKy']"):
        Full_review.append(n.text)
    driver.find_element_by_xpath("//a[@class='_1LKTO3']").click()
    time.sleep(3)
```

```python
#Now we need to check the length of the each columns before creating a DataFrame
```

```python
print(len(Rating_phone),len(Review_summary),len(Full_review))
```

```python
#Now we are going to create a dataFrame for stored data
```

```python
import pandas as pd
```

```python
iphone=pd.DataFrame({})
iphone['Rating_phone']=Rating_phone
iphone['Review_summary']=Review_summary
iphone['Full_review']=Full_review
iphone
```

```python
#lets save it into csv format
```

```python
iphone.to_csv("iphone11_review.csv")
```
